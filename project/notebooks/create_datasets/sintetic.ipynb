```python
import os
import matplotlib.pyplot as plt
plt.style.use('ieee')

from pandas import DataFrame
from sklearn.datasets import make_moons, make_circles
```

```python
root_path = os.path.dirname(os.path.dirname(os.getcwd()))
```

## **Make circles**

```python
# generate 2d classification dataset
fig, axs = plt.subplots(2,2, figsize=(16,8))

X1, y1 = make_circles(n_samples=10000, noise=0.01, factor=0.01, random_state=42) # simples
X2, y2 = make_circles(n_samples=10000, noise=0.1, factor=0.55, random_state=42)
X3, y3 = make_circles(n_samples=10000, noise=0.2, factor=0.25, random_state=42)
X4, y4 = make_circles(n_samples=10000, noise=0.4, factor=0.01, random_state=42)

datas = [(X1, y1), (X2, y2), (X3,y3), (X4, y4)]
for idx, ((X, y), ax) in enumerate(zip(datas, axs.flatten())):
    # scatter plot, dots colored by class value
    df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
    colors = {0:'red', 1:'blue'}
    grouped = df.groupby('label') 
    for key, group in grouped:
        group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key], alpha=0.35)
    ax.set_title(f"Make circles {idx+1}")
    file_path = os.path.join(root_path, "data", "sintetic", f"make_circles_{idx+1}.csv")
    df.to_csv(file_path, index=False)

figure_path = os.path.join(root_path, "images", "make_circles")
plt.savefig(figure_path, facecolor='w')
plt.show()
```

## **Make Moons**
